```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from mpl_toolkits.mplot3d.art3d import Poly3DCollection

# Define vectors
A = np.array([-3, 7, 9])
B = np.array([6, -5, 8])

# Cross product D = A  B
D = np.cross(A, B)

# Cross product C = D  A
C = np.cross(D, A)

# Normalize C to get unit vector
C_unit = C / np.linalg.norm(C)

# Scale C_unit for visibility
C_scaled = C_unit * 5  # Extend it 5x for better visibility

# Plotting
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Origin
origin = np.array([0, 0, 0])

# Plot vectors
ax.quiver(*origin, *A, color='r', label='A', linewidth=2)
ax.quiver(*origin, *B, color='g', label='B', linewidth=2)
ax.quiver(*origin, *D, color='b', label='D = A  B', linewidth=2)
ax.quiver(*origin, *C_scaled, color='purple', label='Unit vector  A in plane', linewidth=2)

# Shade the plane between A and B
plane_AB = [origin, A, B]
poly_AB = Poly3DCollection([plane_AB], alpha=0.3, facecolor='orange')
ax.add_collection3d(poly_AB)

# Shade the plane between D and A
plane_DA = [origin, D, A]
poly_DA = Poly3DCollection([plane_DA], alpha=0.3, facecolor='lightblue')
ax.add_collection3d(poly_DA)

# Set plot limits
ax.set_xlim([-10, 10])
ax.set_ylim([-10, 10])
ax.set_zlim([-10, 10])

# Labels
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Vector Geometry with Shaded Planes')
ax.legend()

plt.show()
```

```python
import numpy as np

A = np.array([-3, 7, 9])
B = np.array([6, -5, 8])
D = np.cross(A, B)
C = np.cross(D, A)
C_unit = C / np.linalg.norm(C)

# Angle between B and C_unit
cos_theta = np.dot(B, C_unit) / (np.linalg.norm(B) * np.linalg.norm(C_unit))
angle_rad = np.arccos(np.clip(cos_theta, -1.0, 1.0))
angle_deg = np.degrees(angle_rad)

print(f"Angle between B and unit vector C: {angle_deg:.2f} degrees")
```
