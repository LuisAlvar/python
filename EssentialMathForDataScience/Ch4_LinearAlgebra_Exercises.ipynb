```python
import numpy as np
import matplotlib.pyplot as plt

# Original vector v
v = np.array([1, 2])

# Transformation matrix: columns are transformed i-hat and j-hat
T = np.array([[2, 0],
              [0, 1.5]])

# Transformed vector v'
v_transformed = T @ v

# Set up subplots
fig, axs = plt.subplots(1, 3, figsize=(15, 5))
titles = ['Original Vector v', 'Transformed Basis Vectors', 'Transformed Vector v']

# Common grid setup
for ax in axs:
    ax.set_xlim(-1, 5)
    ax.set_ylim(-1, 5)
    ax.set_aspect('equal')
    ax.set_xticks(np.arange(-1, 6, 0.5))
    ax.set_yticks(np.arange(-1, 6, 0.5))
    ax.grid(True, which='both', linestyle='--', linewidth=0.5)

# Plot 1: Original vector v
axs[0].quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='blue')
axs[0].set_title(titles[0])
axs[0].annotate(f'v = ({v[0]}, {v[1]})', xy=(v[0], v[1]), xytext=(v[0]+0.2, v[1]), fontsize=12)

# Dashed lines for x and y components
axs[0].plot([0, v[0]], [0, 0], linestyle='--', color='orange')  # x movement
axs[0].plot([v[0], v[0]], [0, v[1]], linestyle='--', color='purple')  # y movement

# Plot 2: Transformed basis vectors
i_hat = T[:, 0]
j_hat = T[:, 1]
axs[1].quiver(0, 0, i_hat[0], i_hat[1], angles='xy', scale_units='xy', scale=1, color='red')
axs[1].quiver(0, 0, j_hat[0], j_hat[1], angles='xy', scale_units='xy', scale=1, color='green')
axs[1].set_title(titles[1])
axs[1].annotate('î → [2, 0]', xy=(i_hat[0], i_hat[1]), xytext=(i_hat[0]+0.2, i_hat[1]), fontsize=12)
axs[1].annotate('ĵ → [0, 1.5]', xy=(j_hat[0], j_hat[1]), xytext=(j_hat[0]+0.2, j_hat[1]), fontsize=12)

# Plot 3: Transformed vector v
axs[2].quiver(0, 0, v_transformed[0], v_transformed[1], angles='xy', scale_units='xy', scale=1, color='purple')
axs[2].set_title(titles[2])
axs[2].annotate(f'T(v) = ({v_transformed[0]}, {v_transformed[1]})', 
                xy=(v_transformed[0], v_transformed[1]), 
                xytext=(v_transformed[0]+0.2, v_transformed[1]), fontsize=12)

# Dashed lines for transformed x and y components
axs[2].plot([0, v_transformed[0]], [0, 0], linestyle='--', color='orange')  # x movement
axs[2].plot([v_transformed[0], v_transformed[0]], [0, v_transformed[1]], linestyle='--', color='purple')  # y movement

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define basis vectors
i_hat = np.array([1, 0])
j_hat = np.array([2, 2])

# Create matrix with basis vectors as columns
M = np.column_stack((i_hat, j_hat))

# Compute determinant
det = np.linalg.det(M)

# Set up subplots
fig, axs = plt.subplots(1, 2, figsize=(12, 5))
titles = ['Basis Vectors', 'Area Representing Determinant']

# Common grid setup
for ax in axs:
    ax.set_xlim(-1, 4)
    ax.set_ylim(-1, 4)
    ax.set_aspect('equal')
    ax.set_xticks(np.arange(-1, 5, 0.5))
    ax.set_yticks(np.arange(-1, 5, 0.5))
    ax.grid(True, linestyle='--', linewidth=0.5)

# Plot 1: Basis vectors
axs[0].quiver(0, 0, i_hat[0], i_hat[1], angles='xy', scale_units='xy', scale=1, color='orange')
axs[0].quiver(0, 0, j_hat[0], j_hat[1], angles='xy', scale_units='xy', scale=1, color='purple')
axs[0].set_title(titles[0])
axs[0].annotate('î = [1, 0]', xy=(i_hat[0], i_hat[1]), xytext=(i_hat[0]+0.2, i_hat[1]), fontsize=12)
axs[0].annotate('ĵ = [2, 2]', xy=(j_hat[0], j_hat[1]), xytext=(j_hat[0]+0.2, j_hat[1]), fontsize=12)

# Plot 2: Parallelogram and determinant
# Define parallelogram corners
origin = np.array([0, 0])
corner1 = i_hat
corner2 = j_hat
corner3 = i_hat + j_hat

# Shade parallelogram
parallelogram = np.array([origin, corner1, corner3, corner2])
axs[1].fill(parallelogram[:,0], parallelogram[:,1], color='lightgrey', alpha=0.6, edgecolor='black')

# Plot vectors again
axs[1].quiver(0, 0, i_hat[0], i_hat[1], angles='xy', scale_units='xy', scale=1, color='orange')
axs[1].quiver(0, 0, j_hat[0], j_hat[1], angles='xy', scale_units='xy', scale=1, color='purple')
axs[1].set_title(titles[1])
axs[1].annotate(f'Determinant = {det:.2f}', xy=(1.5, 1), fontsize=12, backgroundcolor='white')

plt.tight_layout()
plt.show()
```

```python
from numpy import array
from numpy.linalg import inv

# 3x + 1y + 0z = 54
# 2x + 4y + 1z = 12
# 3x + 1y + 8z = 6

# Define the system of equation into a matrix 
A = array([
    [3,1,0],
    [2,4,1],
    [3,1,8]
])

# Define the vector of solutions
B = array([
    54,
    12,
    6
])

# Compute the dot product
X = inv(A).dot(B)

print(X)

from sympy import *

A = Matrix([
    [3,1,0],
    [2,4,1],
    [3,1,8]
])

B = Matrix([
    54,
    12,
    6
])

X = A.inv() * B

print(X)
```
