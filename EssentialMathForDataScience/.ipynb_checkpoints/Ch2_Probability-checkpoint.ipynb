```python
# Using SciPy for the binomial distribution
from scipy.stats import binom 

n = 10
p = 0.9

for k in range(n + 1):
    probability = binom.pmf(k, n, p)
    print("Trial({0}) - {1}".format(k, probability))
```

```python
# Deta distribution using SciPy
from scipy.stats import beta

a = 8 
b = 2

p = beta.cdf(.90, a, b)

print(p)
```

```python
# Subtracting to get a right area in a beta distribution
from scipy.stats import beta

a=8
b=2

p = 1.0 - beta.cdf(0.90, a, b)

print(p)
```

```python
# Example 2-5: A beta distribution with more trials
from scipy.stats import beta
a = 30
b = 6
p = 1.0 - beta.cdf(0.90, a, b)
print(p)
```

```python
# Example 2-6: Beta distrbution middle area using SciPy
a = 8
b = 2
p = beta.cdf(.90, a, b) - beta.cdf(.80, a, b)
print(p)
```

```python
# Additional Example on Binomial Distrubition

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Parameters
n = 10       # trials
p = 0.3      # probability of success
x = np.arange(0, n+1)

# PMF
pmf = binom.pmf(x, n, p)

# Plot
plt.bar(x, pmf, color='skyblue')
plt.title('Binomial Distribution PMF (n=10, p=0.3)')
plt.xlabel('Number of Successes')
plt.ylabel('Probability')
plt.grid(True)
plt.show()
```

```python
# Additional Example: Beta Distribution

from scipy.stats import beta
import matplotlib.pyplot as plt
import numpy as np

# Suppose we observed 8 successes and 4 failures
alpha = 8 + 1
beta_param = 4 + 1

x = np.linspace(0, 1, 100)
pdf = beta.pdf(x, alpha, beta_param)

# Plot
plt.plot(x, pdf, color='purple')
plt.title('Beta Distribution (α=9, β=5)')
plt.xlabel('Probability of Success')
plt.ylabel('Density')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Parameters
n = 137      # trials
p = 0.4      # probability of success
x = np.arange(0, n+1)

# PMF
pmf = binom.pmf(x, n, p)

# Total probability for at least 50 successes
prob_at_least_50 = np.sum(pmf[x >= 50])
print(f"Total probability for at least 50 successes: {prob_at_least_50:.4f}")

# Plot
plt.bar(x, pmf, color='skyblue')
plt.title(f'Binomial Distribution PMF (n={n}, p={p})\nP(X ≥ 50) = {prob_at_least_50:.4f}')
plt.xlabel('Number of Successes')
plt.ylabel('Probability')
plt.grid(True)
plt.show()
```

```python
from scipy.stats import beta
import matplotlib.pyplot as plt
import numpy as np

# Suppose we observed 15 heads and 4 tails
alpha = 15
beta_param = 4

x = np.linspace(0, 1, 100)
pdf = beta.pdf(x, alpha, beta_param)

# Compute area under the curve from x = 0.5 to x = 1
area_50_to_100 = beta.cdf(1, alpha, beta_param) - beta.cdf(0.5, alpha, beta_param)
print(f"Area under the curve from x = 0.5 to x = 1: {area_50_to_100:.4f}")

# Plot
plt.plot(x, pdf, color='purple')
plt.title(f'Beta Distribution (α={alpha}, β={beta_param})\nP(x ≥ 0.5) = {area_50_to_100:.4f}')
plt.xlabel('Probability of Success')
plt.ylabel('Density')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta
import ipywidgets as widgets
from IPython.display import display

# Interactive function
def update_plot(flips=20, heads_ratio=0.5):
    heads = int(flips * heads_ratio)
    tails = flips - heads
    alpha = heads + 1  # add 1 for Bayesian prior
    beta_param = tails + 1

    x = np.linspace(0, 1, 200)
    pdf = beta.pdf(x, alpha, beta_param)

    area = beta.cdf(1, alpha, beta_param) - beta.cdf(0.5, alpha, beta_param)

    plt.figure(figsize=(8, 4))
    plt.plot(x, pdf, color='darkorange', lw=2)
    plt.fill_between(x, pdf, where=(x >= 0.5), color='orange', alpha=0.3)
    plt.title(f'Beta Posterior (Heads={heads}, Tails={tails})\nP(p ≥ 0.5) = {area:.4f}')
    plt.xlabel('Probability of Success')
    plt.ylabel('Density')
    plt.grid(True)
    plt.show()

# Sliders
flips_slider = widgets.IntSlider(value=20, min=10, max=500, step=10, description='Flips')
ratio_slider = widgets.FloatSlider(value=0.5, min=0.1, max=0.9, step=0.05, description='Heads Ratio')

# Display
ui = widgets.VBox([flips_slider, ratio_slider])
out = widgets.interactive_output(update_plot, {'flips': flips_slider, 'heads_ratio': ratio_slider})
display(ui, out)
```
